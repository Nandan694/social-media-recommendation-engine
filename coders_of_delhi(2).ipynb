```python
import json
def clean_data(data):
    # remove empty names
    data["users"] = [user for user in data["users"] if user["name"].strip()] # user for user in data['users'] means all the users inside data
    # if user["name"].strip() removes any empty names in all the users inside 'data'
    
    # remove duplicate friends
    for user in data['users']: """ user- temporarily creating it to store data['users'], and innside user of friends, using set to remove
                                                              duplicate friends and then convert it back to list """
        user['friends'] = list(set(user['friends'])) 

    #remove inactive users
    data["users"] = [user for user in data["users"] if user['friends'] or user['liked_pages']]
    # same as the 1st one, but difference is using OR, if either one of them is true, print it, if both are false discard it

    #remove duplicate pages
    unique_pages = {}
    for page in data['pages']:
        unique_pages[page['id']] = page
    data['pages'] = list(unique_pages.values())
    return data

data = json.load(open("data2.json")) # loading the data to data2.json directly
data = clean_data(data) # calling the function
json.dump(data, open("cleaned_data2.json","w"), indent=4) # dumping the refined data to a new file, by opening a new file in write mode
print("Data has been cleaned successfully") # just for readability
```
