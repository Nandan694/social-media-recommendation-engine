```python
import json
```

```python
def load_data(filename): # to load data
    with open(filename, "r") as f: # storing filename in read mode in 'f'
        return json.load(f) # to read json file that's in f

def find_people_you_may_know(user_id, data): # user_id - the id of who we want to display reccomendations to him, data- massive_data.json
    user_friends = {} # making an empty dictionary 
    for user in data['users']: # temporarily storing data['users'] in user..
        user_friends[user['id']] = set(user['friends']) # to get only unique values corresponding to user_id 

    if user_id not in user_friends: # data sanitization
        return [] # considering the case where the entered user_id does not exist in the data

    direct_friends = user_friends[user_id] # direct friends of the user who we want to recommend the members who shares the mutual friends
    suggestions = {} 
    for friend in direct_friends: # if we take '1' as user_id, direct_friends - 2,3, running loop on both 2 and 3 
        for mutual in user_friends[friend]: # friends's friend
            if mutual!=user_id and mutual not in direct_friends: # considering the case where the system does not recommend his or direct_friend's id
                # core logic  #more efficient than +=1, takes the existed and repeated ones and increments it
                suggestions[mutual] = suggestions.get(mutual, 0) + 1 # if its totally new, starts adding from one, better use case for large databases
    sorted_suggestions = sorted(suggestions.items(), key = lambda x: x[1], reverse = True) # taking the dictionary's items and key,
    # using reverse for to give the one's who has been on mutual friend's list for most, like repeating the most times, since sorting does
    # in ascending to descending order, so use of reverse, for to write the order in descending to ascending order
    return [user_id for user_id, mutual_count in sorted_suggestions] # all the user_id's and the total mutual count inside the sorted_suggestions

# simple and easy to understand, no explanations are required
data = load_data("massive_data.json")
user_id = 10
recc = find_people_you_may_know(user_id, data)
print(recc)
```
